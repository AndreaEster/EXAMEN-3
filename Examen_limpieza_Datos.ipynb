```python
dfq1 = df.loc[ : , ["Months_Inactive_12_mon","Credit_Limit","CLIENTNUM"] ].groupby( ["Months_Inactive_12_mon","Credit_Limit"] , as_index = False ).quantile(0.25)
dfq1.columns = ["Months_Inactive_12_mon","Credit_Limit","CLIENTNUM"]
```

```python
dfq3 = df.loc[ : , ["Months_Inactive_12_mon","Credit_Limit","CLIENTNUM"] ].groupby( ["Months_Inactive_12_mon","Credit_Limit"] , as_index = False ).quantile(0.75)
dfq3.columns = ["Months_Inactive_12_mon","Credit_Limit","CLIENTNUM_3"]
```

```python
dfq2 = df.loc[ : , ["Months_Inactive_12_mon","Credit_Limit","CLIENTNUM"] ].groupby( ["Months_Inactive_12_mon","Credit_Limit"] , as_index = False ).quantile(0.50)
dfq2.columns = ["Months_Inactive_12_mon","Credit_Limit","CLIENTNUM_2"]
```

```python
dfq1
```

```python
dfq3
```

```python
df_iqr = dfq1.merge( dfq3 , on=["Months_Inactive_12_mon","Credit_Limit"] , how="inner" ).merge( dfq2 , on=["Months_Inactive_12_mon","Credit_Limit"] , how="inner" )
```

```python
df_iqr
```

```python
df_iqr["iqr"] = df_iqr.apply( lambda r: r["CLIENTNUM_3"] - r["CLIENTNUM"] , axis = 1)
```

```python
df_iqr
```

```python
df_iqr["upper"] = df_iqr.apply( lambda r: ( r["CLIENTNUM_3"] + 1.5*r["iqr"] ) , axis = 1)
df_iqr["lower"] = df_iqr.apply( lambda r: ( r["CLIENTNUM"] - 1.5*r["iqr"] ) , axis = 1)
```

```python
df_iqr
```

```python
df_t = df.merge( df_iqr , on=["Months_Inactive_12_mon","Credit_Limit"] , how="left" ).copy()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("BankChurners.csv")
df
```

```python
df.columns
```

```python
df.dtypes
```

```python
cols = ['CLIENTNUM',
       'Income_Category', 'Card_Category', 'Months_on_book',
       'Total_Relationship_Count', 'Months_Inactive_12_mon',
       'Contacts_Count_12_mon', 'Credit_Limit', 'Total_Revolving_Bal',
       'Avg_Open_To_Buy', 'Total_Amt_Chng_Q4_Q1', 'Total_Trans_Amt',
       'Total_Trans_Ct', 'Total_Ct_Chng_Q4_Q1', 'Avg_Utilization_Ratio']
```

```python
cols_validas = []
cols_many_levels = []
cols_one_level = []

size_df = df.shape[0]
umbral = .95

for c_ in cols:
    
    niveles = df.loc[ : , [c_] ].drop_duplicates().shape[0]
    
    if ( niveles / size_df ) >= umbral:
        cols_many_levels.append(c_)
    elif niveles == 1:
        cols_one_level.append(c_)
    else:
       cols_validas.append(c_)
```
